### Handling missing values in categorical features

```python
# Filtering the categorical features that has more than 1 nan value
categorical_features_nan = [features for features in housing.columns if housing[features].isnull().sum()>1 and housing[features].dtype== 'O']
```

```python
categorical_features_nan
```

```python
def replace_nan_cat_features(housing,categorical_features_nan):
  data = housing.copy()
  data[categorical_features_nan] = data[categorical_features_nan].fillna('Missing')
  return data
housing = replace_nan_cat_features(housing,categorical_features_nan)
housing.head()
```

### Handling Missing Values for Numerical features

```python
numerical_nan_features = [features for features in housing.columns if housing[features].dtypes != 'O']
numerical_nan_features
```

```python
## Replacing numerical missing values with median since there are outliers.
for feature in numerical_nan_features:
    median_value=housing[feature].median()
    housing[feature+'nan']=np.where(housing[feature].isnull(),1,0)
    housing[feature].fillna(median_value,inplace=True)
    
housing[numerical_nan_features].isnull().sum()
```

```python
## Temporal Variables
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       housing[feature]=housing['YrSold']-housing[feature]
housing.head()
```

### Performing log normal distribution on numerical features since the distribution is skewed

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    housing[feature]=np.log(housing[feature])
```

### Handling rare categorical features

```python
categorical_nan_features = [features for features in housing.columns if housing[features].dtypes == 'O']
categorical_features_nan
```

```python
# Replacing the rare categorical feature values with the valu rare.
for feature in categorical_nan_features:
    temp=housing.groupby(feature)['SalePrice'].count()/len(housing)
    temp_df=temp[temp>0.01].index
    housing[feature]=np.where(housing[feature].isin(temp_df),housing[feature],'Rare_var')
    
```

```python
# Mapping categorical features
for feature in categorical_nan_features:
    labels_ordered=housing.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    housing[feature]=housing[feature].map(labels_ordered)
```

```python
housing.head()
```

### Feature scaling 

```python
feature_scale=[feature for feature in housing.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(housing[feature_scale])
```

```python
scaler.transform(housing[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([housing[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(housing[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.to_csv('X_train.csv',index=False)
```

## Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
Y_train = data[['SalePrice']]
Y_train
```

```python
## Dropping ID and Saleprice column from train dataset
X_train = data.drop(['Id','SalePrice'], axis=1)
X_train.head()
```

```python
# Using Lasso regression and selectfrommodel  for feature selection
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) 
feature_sel_model.fit(X_train, Y_train)
```

```python
selected_features = X_train.columns[(feature_sel_model.get_support())]
selected_features
```
